```python
import pandas as pd
```

```python
data=pd.read_csv("D:\\Downloads\\titanic_data.csv")
```

```python
data.isnull().sum()
```

```python
data1=data.drop('Cabin',axis=1)
```

```python
data1.shape
```

```python
data1.dropna(axis=0,inplace=True)
```

```python
data1.head()
```

```python
pd.get_dummies(data1[['Sex','Embarked']])
```

```python
dummied_data=pd.get_dummies(data1[['Sex','Embarked']])
```

```python
data1=data.drop(['Sex','Embarked'],axis=1)
```

```python
datafl=data1.merge(dummied_data,left_index=True,right_index=True)
```

```python
datafl.head()
```

```python
datafl.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=datafl[['PassengerId','Pclass','Age','SibSp','Parch','Sex_female','Sex_male','Embarked_C','Embarked_Q','Embarked_S']]
```

```python
y=datafl['Survived']
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(train_x,train_y)
```

```python
pred_y=model.predict(test_x)
```

```python
result=pd.DataFrame({'actual':test_y,'predicted':pred_y})
```

```python
print(result)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,precision_score,recall_score,f1_score,roc_curve
```

```python
accuracy_score(test_y,pred_y)
```

```python
confusion_matrix(test_y,pred_y)
```

```python
precision_score(test_y,pred_y)
```

```python
recall_score(test_y,pred_y)
```
