```python
import pandas as pd
```

```python
data=pd.read_csv("D:\\Downloads\\rainfall.csv")
```

```python
data.isnull().sum()
```

```python
data['NOV']=data['NOV'].fillna((data['NOV'].mean()))
```

```python
data['JUL']=data['JUL'].fillna((data['JUL'].mean()))
```

```python
data[['APR','JUN','AUG','SEP','OCT','DEC']]=data[['APR','JUN','AUG','SEP','OCT','DEC']].fillna((data[['APR','JUN','AUG','SEP','OCT','DEC']].mean()))
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
data['STATE_UT_NAME']=le.fit_transform(data['STATE_UT_NAME'])
```

```python
data['DISTRICT']=le.fit_transform(data['DISTRICT'])
```

```python
data1=data['NOV']
```

```python
data.drop(['NOV','ANNUAL','Jan-Feb','Mar-May','Jun-Sep','Oct-Dec'],axis=1,inplace=True)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=data
```

```python
y=data1
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(train_x,train_y)
```

```python
y_pred=model.predict(test_x)
```

```python
pd.DataFrame({'actual':test_y,'predict':y_pred})
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
print(r2_score(test_y,y_pred))
```

```python
print(mean_squared_error(test_y,y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor( n_estimators=50,min_samples_split=10,min_samples_leaf=5)
```

```python
model.fit(train_x,train_y)
```

```python
y_pred=model.predict(test_x)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
print(r2_score(test_y,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model=RandomForestRegressor()
```

```python
param_grid={'n_estimators':[50,100,150],'min_samples_split':[10,20,30],'max_features':[4,5,7]}
```
